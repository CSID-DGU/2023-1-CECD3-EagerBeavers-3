```python
import ast
import networkx as nx
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from google.colab import drive
import os

# # Google Drive 마운트
# drive.mount('/content/drive')

# Google Drive에 올린 파일의 경로
file_path = '/content/drive/MyDrive/Project_CodeNet_Python800/p00000/'

files = os.listdir(file_path)
# print(files)

python_codes = []

for file_name in files:

    directory = os.path.join(file_path, file_name)

    # 파일을 열기
    with open(directory, 'r') as file:
      code = file.read()
      python_codes.append(code)

# for code in python_codes :
#   print(code)


def code_to_ast(code):
    return ast.parse(code)

def ast_to_tokens(node):
    tokens = []
    for child_node in ast.iter_child_nodes(node):
        tokens.append(type(child_node).__name__)
        tokens.extend(ast_to_tokens(child_node))
    return tokens



# 각 코드를 AST로 변환하고 토큰화하여 문자열로 변환
code_strs = []
for code in python_codes:
  print("코드:"+code)
  print("############################")
  try :
    parsed_ast = code_to_ast(code)
    print("@@@@@@@@@처리 성공@@@@@@@@@@")
  except :
    print("@@@@@@@@처리 실패@@@@@@@@@@")
    continue


  tokens = ast_to_tokens(parsed_ast)
  code_str = ' '.join(tokens)
  code_strs.append(code_str)


# print(code_strs)

print("@@@@@@@@@@@@@@@@@@@@@")
print("ast화, 토큰화 완료")
print("@@@@@@@@@@@@@@@@@@@@@")

print("@@@@@@@@@@@@@@@@@@@@@")
print("code strs")
print(code_strs)
print("@@@@@@@@@@@@@@@@@@@@@")

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(code_strs)

# 정규화
normalized_matrix = normalize(tfidf_matrix)

# K-means 클러스터링 (n_samples >= n_clusters)
n_clusters = 5  # 클러스터의 수
n_samples = len(code_strs)  # 코드 샘플의 개수로 설정
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(normalized_matrix)

# 클러스터링 결과
print("Cluster labels:", kmeans.labels_)
```
